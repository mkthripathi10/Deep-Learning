```python
# multiple images in single frames veritcal and vorizontal
```

```python
imgD = cv.imread("img.jpg")
imgD1 = cv.imread("dog.jpg")
resize = cv.resize(imgD,(200,200))
resize1 = cv.resize(imgD1,(200,200))

horiz = np.hstack([resize,resize1,resize,resize1])
vertical = np.vstack([horiz,horiz,horiz])
cv.imshow("img",vertical)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# slide the images
```

```python
imgList = ["dog.jpg","dog1.jpg","img.jpg"]
for i in imgList:
    img = cv.imread(i,1)
    resize = cv.resize(img,(500,300))
    cv.imshow("Manish",resize)
    cv.waitKey(0)
    
cv.destroyAllWindows()    
```

```python
# overlap the images
```

```python
img = cv.imread("dog.jpg")
img1 = cv.imread("dog1.jpg")

res = cv.resize(img,(400,300))
res1 = cv.resize(img1,(400,300))

overLap = cv.addWeighted(res,2,res1,0.7,4)
cv.imshow("overlap",overLap)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#edge detection
```

```python
img = cv.imread("dog.jpg")
res = cv.resize(img,(400,300))
newImg = cv.Canny(res,100,70)
cv.imshow("Img",newImg)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# blur the images
```

```python
img = cv.imread("dog2.jpg",1)
if img is None:
    print("no image")
else:
    res = cv.resize(img,(300,200))
#     newImg = cv.medianBlur(res,5)
    newImg = cv.bilateralFilter(res,12,100,100)
    cv.imshow("Img",newImg)
    cv.waitKey(0)
    cv.destroyAllWindows()
```

```python
# save images
```

```python
listG=[]
dogList = ["dog.jpg","dog1.jpg","dog2.jpg"]
for i in dogList:
    if i is None:
        print("No Image")
    else:
        read = cv.imread(i)
        resize = cv.resize(read,(300,200))
        listG.append(resize)
    
horiz = np.hstack(listG)
cv.imwrite("allDog.jpg",horiz)
cv.imshow("window",horiz)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# capture video from webcam
```

```python
video = cv.VideoCapture(0)

while video.isOpened():
    ret,frame = video.read()
    if ret == True:
        cv.imshow("opne",frame)
        if cv.waitKey(25) & 0xff == ord("p"):
            break
    else:
        break
   
        
video.release()
cv.destroyAllWindows()
```

```python
# Read video from the file
```

```python
cap = cv.VideoCapture("AWS-Solutions.mp4")
while cap.isOpened():
    ret, fram = cap.read()
    print(ret,"-----",fram,"-----",fram.shape)
    if ret == True:
        fram =cv.resize(fram,(600,500))
        cv.imshow("web",fram)
        if cv.waitKey(25) & 0xff == ord('p'):
            break   
    else:
        break
        
cap.release()
cv.destroyAllWindows()
```

```python
# fast motion and slow motion
```

```python
vid = cv.VideoCapture(0)
while vid.isOpened():
    r,frame =vid.read()
    if r == True:
        frame = cv.resize(frame,(500,400))
        cv.imshow("motion",frame)
        if cv.waitKey(8) & 0xff == ord("p"):
            break
            
    else:
        break
        
vid.release()
cv.destroyAllWindows()
```
