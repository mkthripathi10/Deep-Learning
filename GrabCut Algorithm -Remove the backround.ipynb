```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load image

img = cv.imread("dog1.jpg")
if img is None:
    print("No image found")
else:
    mask = np.zeros(img.shape[:2],np.uint8)
    
    #Temporary  arrarys used by the alogrithm
    backgroundMask = np.zeros((1,65),np.float64)
    forgroundMask = np.zeros((1,65),np.float64)
    
    #rectangle around the object (x,y,w,h)
    rec = (50,50,450,290)
    
    #Apply GrabCut Alogrith
    cv.grabCut(img,mask,rec,backgroundMask,forgroundMask,10,cv.GC_INIT_WITH_RECT)
    
    #convert to mask to binary
    mask2 = np.where((mask ==2) | (mask ==0),0,1).astype('uint8')
    
    img_result = img * mask2[:,:,np.newaxis]
    print(img_result)
    
    #display reslut
    image = cv.cvtColor(img_result, cv.COLOR_BGR2RGB)

    plt.imshow(image)
    plt.axis("off")
    plt.title("remove")
    plt.show()
    
```
